```python
Calls= pd.read_csv("FactCallCenter.csv")
```

```python
Calls['ServiceGrade']=Calls['ServiceGrade'].astype(int)
```

```python
Calls.describe()
```

```python
#For the EDA
#Create a scatterplot of Calls versus Orders made where the points are colored by the Shift column. **
```

```python
sns.set_style('whitegrid')
sns.lmplot('Calls','Orders',data=Calls, hue='Shift',
           palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
#Create a scatterplot of orders versus total operators available where the points are colored by the Shiftcolumn.
```

```python
sns.set_style('whitegrid')
sns.lmplot('Orders','TotalOperators',data=Calls, hue='Shift',
           palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
#Create a stacked histogram showing number of orders made.
```

```python
sns.set_style('darkgrid')
g = sns.FacetGrid(Calls,hue="Shift",palette='coolwarm',size=6,aspect=2)
g = g.map(plt.hist,'Orders',bins=20,alpha=0.7)
```

```python
#Create a similar histogram for calls.
```

```python
sns.set_style('darkgrid')
g = sns.FacetGrid(Calls,hue="Shift",palette='coolwarm',size=6,aspect=2)
g = g.map(plt.hist,'Calls',bins=20,alpha=0.7)
```
